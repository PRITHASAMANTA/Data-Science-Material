```python
import pandas as pd
```

```python
df=pd.read_csv('/content/titanic.csv')
```

```python
df.head()
```

```python
#inspect data
print(df.info())
```

```python
df.describe()
```

```python
#handle missing values
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
#remove duplicates

df=df.drop_duplicates()
```

```python
#filter data :passenger in first class
first_class=df[df['Pclass']==1]
```

```python
print("First Class Passengers:")
print(first_class.head())
```

# Task 2: Generate visualizations to illustrate key insights

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#display barchart: survival rate by class
survival_by_class = df.groupby('Pclass')['Survived'].mean()
plt.bar(survival_by_class.index, survival_by_class.values)
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
#Age distribution
sns.histplot(df["Age"],kde= True,bins=20,color='red')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

```python
#Scatter Plot: Age vs Fare
plt.scatter(df['Age'], df['Fare'], alpha=0.5,color='blue')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')
plt.show()
```
